```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

file_paths = [
    "As_jr.txt",
    "P_jr.txt"
]

fig, ax = plt.subplots(figsize=(2.4, 2.4))
label=[r'$a$-As', r'$a$-P']
colors=['#C17F91', '#7c87b5'] 
scale_factor = [2.487, 2.232]

for i, file_path in enumerate(file_paths):
    data = pd.read_csv(file_path, delim_whitespace=True, skiprows=1, header=None)
    x = data[0] / scale_factor[i]
    y = data[1]
    if i == 0:
        plt.plot(x, y, label=label[i], color=colors[i], linestyle='-', zorder=4)
    else:
        plt.plot(x, y, label=label[i], color=colors[i], linestyle='--', dashes=[2, 1], zorder=2)

ax.set_xlabel(r"Scaled distance")
ax.set_ylabel(r"${J}$(${r}$) (atoms $\mathrm{\AA^{-1}}$)")
ax.set_xlim(0, 2.75)
ax.set_xticks(np.arange(0, 2.75, 0.5))
ax.set_ylim(-2, 30)
ax.set_yticks(np.arange(0, 30.1, 10))
ax.legend()
ax.yaxis.set_tick_params(which='both', labelbottom=True, bottom=True, zorder=10)
ax.xaxis.set_tick_params(which='both', labelbottom=True, bottom=True, zorder=10)

plt.show()
```
